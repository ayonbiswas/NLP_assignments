```python
import nltk
```

importing Tom Sawyer Corpus

```python
file = open("./74-0.txt",'r')
```

```python
text = file.read()
```

compute tokens and removing punctuations

```python
tokens = nltk.word_tokenize(text)
#remove punctuations
tokens = [word for word in tokens if word.isalpha()]
print (len(tokens))
```

compute types

```python
types = set(tokens)
print(types)
```

compute TTR

```python
TTR = float(len(types))/float(len(tokens))
print("TTR: {}".format(TTR))
```

plotting Zipf's law

```python
frequency = dict()
for word in tokens:
    count = frequency.get(word,0)
    frequency[word] = count + 1
```

```python
from operator import itemgetter
rank = 1
prod = []
for key, value in reversed(sorted(frequency.items(), key = itemgetter(1))):
    print ("rank: {}, token :{}, value :{}, product: {} ".format(rank,key, value, value*rank ))
    prod.append(value*rank)
    rank+=1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

plotting product of rank and frequency Vs rank

```python
plt.plot(range(len(prod)), prod)
plt.xlabel("rank")
plt.ylabel("product of rank and frequency")
plt.title("Zipf's Law")
plt.grid()
plt.ylim(0,15000)
```

Finding parameters of Heap's Law

```python
def heap_law(x,k,b):
    y = k*(x**b)
    return y
```

```python
from scipy.optimize import curve_fit
param, covar = curve_fit(heap_law, x, y)
print(param)
```
